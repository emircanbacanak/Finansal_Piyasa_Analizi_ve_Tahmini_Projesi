# Özellik Mühendisliği
**Veri setinde hisse senedi fiyat tahminlerini iyileştirebilecek özellikler oluşturulmuştur. Bu özellikler şunlardır:**

**Open (Açılış Fiyatı):** Günlük açılış fiyatı.

**High (En Yüksek Fiyat):** Günlük en yüksek fiyat.

**Low (En Düşük Fiyat):** Günlük en düşük fiyat.

**Volume (Hacim):** Günlük işlem hacmi.

**Returns (Getiriler):** Günlük kapanış fiyatlarına göre hesaplanan yüzdelik değişim.

**Özellik mühendisliği işlemi**

```python
stock_data['Returns'] = stock_data['Close'].pct_change()
stock_data.dropna(inplace=True)
```

# Veri Bölme


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X = stock_data[['Open', 'High', 'Low', 'Volume', 'Returns']]
y = stock_data['Close'].values.reshape(-1, 1)

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

# Model Eğitimi ve Tahmin (Random Forest Regressor)


```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search_rf = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=5, n_jobs=-1)
grid_search_rf.fit(X_train, y_train.ravel())
best_rf = grid_search_rf.best_estimator_
```

# Performans Değerlendirmesi


```python
from sklearn.metrics import mean_squared_error, r2_score

# Test veri seti üzerinde tahmin yap
y_pred_rf_scaled = best_rf.predict(X_test)
y_pred_rf = scaler_y.inverse_transform(y_pred_rf_scaled.reshape(-1, 1))

mse_rf = mean_squared_error(scaler_y.inverse_transform(y_test), y_pred_rf)
r2_rf = r2_score(scaler_y.inverse_transform(y_test), y_pred_rf)

print(f'Mean Squared Error: {mse_rf}')
print(f'R-squared: {r2_rf}')
print("Random Forest Best Parameters:", grid_search_rf.best_params_)
```

# Tahmin Sonuçlarının Görselleştirilmesi
**Tahmin edilen fiyatlar ile gerçek fiyatlar karşılaştırılmış ve grafiklerle görselleştirilmiştir:**

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x=scaler_y.inverse_transform(y_test).flatten(), y=y_pred_rf.flatten())
plt.plot([min(scaler_y.inverse_transform(y_test).flatten()), max(scaler_y.inverse_transform(y_test).flatten())],
         [min(scaler_y.inverse_transform(y_test).flatten()), max(scaler_y.inverse_transform(y_test).flatten())], color='red', lw=2)
plt.xlabel('Gerçek Değerler')
plt.ylabel('Tahmin Edilen Değerler')
plt.title('Gerçek Değerler vs Tahmin Edilen Değerler (Random Forest)')
plt.show()
```

**artıkların (residuals) histogramını ve KDE (Kernel Density Estimate) grafiği**

```python
residuals_rf = scaler_y.inverse_transform(y_test) - y_pred_rf
plt.figure(figsize=(10, 6))
sns.histplot(residuals_rf, kde=True)
plt.xlabel('Artıklar (Gerçek - Tahmin)')
plt.title('Artıkların Dağılımı (Random Forest)')
plt.show()
```

#Random Forest Regressor ile Haftalık Tahmin


```python
last_observation = stock_data.iloc[-1]
new_data_week = [[last_observation['Open'], last_observation['High'], last_observation['Low'], last_observation['Volume'], last_observation['Returns']]]
new_data_week_scaled = scaler_X.transform(new_data_week)
predicted_close_price_week_rf = scaler_y.inverse_transform(best_rf.predict(new_data_week_scaled).reshape(-1, 1))
print(f"Bir hafta sonrası tahmini kapanış fiyatı (Random Forest): {predicted_close_price_week_rf[0][0]}")
```

# Model Eğitimi ve Tahmin (Gradient Boosting Regressor)


```python
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5, 7]
}

grid_search_gb = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid_gb, cv=5, n_jobs=-1)
grid_search_gb.fit(X_train, y_train.ravel())
best_gb = grid_search_gb.best_estimator_
```

# Portföy Performansı ve Backtest

```python
import vectorbt as vbt
import pandas as pd
```

**Haftalık Tahmin**

```python
initial_balance = 1000
balance = initial_balance
position = 0
cash = initial_balance
weekly_balance = []

test_data = stock_data.iloc[len(stock_data) - len(X_test):]

for i in range(0, len(test_data) - 7, 7):
    current_data = test_data.iloc[i:i+7]
    if len(current_data) < 7:
        break
    current_close = current_data['Close'].iloc[-1]
    X_week = current_data[['Open', 'High', 'Low', 'Volume', 'Returns']]
    X_week_scaled = scaler_X.transform(X_week)

    y_pred_rf = best_rf.predict(X_week_scaled)
    y_pred_gb = best_gb.predict(X_week_scaled)
    predicted_close_next_week_rf = scaler_y.inverse_transform(y_pred_rf.reshape(-1, 1))[-1]
    predicted_close_next_week_gb = scaler_y.inverse_transform(y_pred_gb.reshape(-1, 1))[-1]

    # Haftalık işlem stratejisi
    if predicted_close_next_week_rf > current_close and cash > 0:
        # Alım sinyali
        position = cash / current_close
        cash = 0
    elif predicted_close_next_week_rf < current_close and position > 0:
        # Satım sinyali
        cash = position * current_close
        position = 0

    weekly_balance.append(cash + position * current_close)
```

**VectorBT ile Portföy Oluşturma**

```python
buy_signals = test_data['Close'].shift(-1) > test_data['Close']
sell_signals = test_data['Close'].shift(-1) < test_data['Close']
portfolio = vbt.Portfolio.from_signals(
    close=test_data['Close'],
    entries=buy_signals,
    exits=sell_signals,
    init_cash=initial_balance,
    fees=0.001
)
```

**Backtest sonuçlarını görselleştirme**

```python
portfolio.plot().show()
```

**Performans Metrikleri**

```python
print(portfolio.stats())
```
